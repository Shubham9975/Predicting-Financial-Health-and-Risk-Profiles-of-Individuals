```python
# importing needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# split data in feature and target
X = train_df.iloc[:, :-1]
y = train_df['Risk Profile']
```

```python
# split dataset in train and test
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=27)
```

```python
# random forest model
forest_model = RandomForestClassifier(n_estimators=100, random_state=27)
```

```python
# building the model
forest_model.fit(X_train, y_train)
```

```python
y_pred = forest_model.predict(X_test)
```

```python
# Evaluating the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# hyper parameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
```

```python
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=27),
                           param_grid=param_grid,
                           cv=3,
                           n_jobs=-1,
                           verbose=2)
```

```python
grid_search.fit(X_train, y_train)
```

```python
# best model after hyper parameter tuning
print("Best Parameters:", grid_search.best_params_)
```

```python
forestCV_model = grid_search.best_estimator_
```

```python
y_pred_forestCV = forestCV_model.predict(X_test)
```

```python
# Evaluating model
print("Accuracy Score (Tuned):", accuracy_score(y_test, y_pred_forestCV))
print("\nConfusion Matrix (Tuned):")
print(confusion_matrix(y_test, y_pred_forestCV))
print("\nClassification Report (Tuned):")
print(classification_report(y_test, y_pred_forestCV))
```

```python
# taking top features to built model
top_features = [
    'Mortgage Information',
    'Savings Rate',
    'Inflation Rate',
    'Emergency Fund Status',
    'Debt-to-Income Ratio',
    'Local Unemployment Rate',
    'Investment Accounts',
    'Credit Card Usage',
    'Interest Rates',
    'Average Monthly Expenses',
    'Credit Score',
    'Age',
    'Bank Account Activity',
    'Years in Current Job',
    'Number of Credit Inquiries'
]
```

```python
X = train_df[top_features]
y = train_df['Risk Profile']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=27)
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5],
    'criterion': ['gini', 'entropy']
}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(random_state=27),
                           param_grid,
                           cv=3,
                           scoring='accuracy',
                           verbose=2,
                           n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model
```

```python
y_pred = best_model.predict(X_test)
```

```python
print("Tuned Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix (Tuned):\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report (Tuned):\n", classification_report(y_test, y_pred))
```
